```python
print("═" * 50)
print("   Oi! Wanna know your BMI real quick?".center(50))
print("═" * 50)

while True:
    while True:
        stuff = input("\nHow heavy are you in kg? (like 67 or 82.3) → ").strip()
        try:
            weight = float(stuff.replace(",", "."))
            if weight > 0 and weight < 400:
                break
            print("bruh that number is wild, try again")
        except:
            print("nah i need actual digits my friend")

    while True:
        tall = input("Height in meters? (1.70 for 170cm etc) → ").strip()
        try:
            height = float(tall.replace(",", "."))
            if 1.20 < height < 2.30:
                break
            print("you sure that's meters and not cm? lol")
        except:
            print("that ain't a height i recognise chief")

    bmi = weight / (height * height)

    print("\n" + "—" * 50)
    print("gimme a sec… doing wizard math")
    print("—" * 50)

    if bmi < 16:
        print(f"→ {bmi:.1f} … mate that's proper skinny")
        print("go see a doc yeah? no joke")
    elif bmi < 17:
        print(f"→ {bmi:.1f} still very light")
        print("maybe grab an extra sandwich or ten")
    elif bmi < 18.5:
        print(f"→ {bmi:.1f} bit on the slim side")
        print("you're good but could eat more pizza")
    elif bmi < 25:
        print(f"→ {bmi:.1f} absolute gold zone!")
        print("whatever you're doing, never stop")
    elif bmi < 30:
        print(f"→ {bmi:.1f} carrying a lil extra")
        print("few walks and you're back in business")
    elif bmi < 35:
        print(f"→ {bmi:.1f} getting chunky")
        print("time to love yourself with some movement")
    elif bmi < 40:
        print(f"→ {bmi:.1f} proper heavy now")
        print("your body would love some help")
    else:
        print(f"→ {bmi:.1f} that's a big number")
        print("doctor time, for real, take care of you")

    print("\n" + "═" * 50)
    again = input("Another go? (just type y or hit enter to bounce) → ").strip().lower()
    if again not in ["y", "yes", "yeah", "yep", "ok", "sure", "go", "1"]:
        print("\nLater dude — stay awesome!")
        break
    print("\nHell yeah, round two!\n")
```
